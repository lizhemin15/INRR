```python
from MinPy import task
import torch as t
from MinPy.toolbox.plot import plot_tra
t.autograd.set_detect_anomaly(True)

para = [240,240,240,240]
para = [240,240]
data_path='./train_pics/Cameraman.jpg'
mask_path='./masks/mask.png'
mask_mode='random'
random_rate=0.75
reg_mode='TV' #'TV',''AIR','NN',None
std_w=1e-3
epoch=2001
model_name = 'fourier'

task_dmf = task.shuffle_task(m=para[0],n=para[1],data_path=data_path,mask_mode=mask_mode,random_rate=random_rate,
                     mask_path=mask_path,given_mask=None,para=para,input_mode='masked',
                    std_b=1e-1,reg_mode=reg_mode,model_name='fp',pro_mode='mask',
                                 opt_type='Adam',shuffle_mode='I',verbose=False,std_w=std_w)
task_dmf.train(epoch=epoch,print_epoch=200,imshow_epoch=200,plot_mode='gray',
                   verbose=False,imshow=True,stop_err=1e-4)
plot_tra(pro_list=task_dmf.pro_list,color='r',label='MSE',pointsize=10)
```
